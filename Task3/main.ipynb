```python
corpus1 = nltk.sent_tokenize(artikel_1.lower())
for i in range(len(corpus1 )):
    corpus1 [i] = corpus1 [i].lower()
    corpus1 [i] = re.sub(r'\W',' ',corpus1 [i])
    corpus1 [i] = re.sub(r'\s+',' ',corpus1 [i])
    print(i.__str__() + ". " + corpus1[i])

corpus2 = nltk.sent_tokenize(artikel_2.lower())
for i in range(len(corpus2 )):
    corpus2 [i] = corpus2 [i].lower()
    corpus2 [i] = re.sub(r'\W',' ',corpus2 [i])
    corpus2 [i] = re.sub(r'\s+',' ',corpus2 [i])
    print(i.__str__() + ". " + corpus2[i])

wordfreq = {}
for sentence in corpus1:
    tokens = nltk.word_tokenize(sentence)
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1

wordfreq2 = {}
for sentence2 in corpus2:
    tokens2 = nltk.word_tokenize(sentence2)
    for token2 in tokens2:
        if token2 not in wordfreq2.keys():
            wordfreq2[token2] = 1
        else:
            wordfreq2[token2] += 1

import heapq
most_freq = heapq.nlargest(200, wordfreq, key=wordfreq.get)
print(corpus1)
most_freq2 = heapq.nlargest(200, wordfreq2, key=wordfreq2.get)
print(corpus2)
```

Menghitung nilai dari idf


```python
word_idf_values1 = {}
for token in most_freq:
    doc_containing_word = 0
    for document in corpus1:
        if token in nltk.word_tokenize(document):
            doc_containing_word += 1
            
    word_idf_values1[token] = np.log10(len(corpus1)/(doc_containing_word))

word_idf_values2 = {}
for token2 in most_freq2:
    doc_containing_word = 0
    for document2 in corpus2:
        if token2 in nltk.word_tokenize(document2):
            doc_containing_word += 1

            
    word_idf_values2[token2] = np.log10(len(corpus2)/(doc_containing_word))
```

Menghitung nilai dari tf

```python
word_tf_values1 = {}
for token in most_freq:
    sent_tf_vector = []
    for document in corpus1:
        doc_freq = 0
        for word in nltk.word_tokenize(document):
            if token == word:
                  doc_freq += 1
        word_tf = doc_freq # tanpa normalisasi
        #word_tf = doc_freq/len(nltk.word_tokenize(document)) # normalisasi dengan panjang dokumen
        sent_tf_vector.append(word_tf)
    word_tf_values1[token] = sent_tf_vector

word_tf_values2 = {}
for token2 in most_freq2:
    sent_tf_vector = []
    for document2 in corpus2:
        doc_freq = 0
        for word in nltk.word_tokenize(document2):
            if token2 == word:
                  doc_freq += 1
        word_tf = doc_freq # tanpa normalisasi
        #word_tf = doc_freq/len(nltk.word_tokenize(document)) # normalisasi dengan panjang dokumen
        sent_tf_vector.append(word_tf)
    word_tf_values2[token2] = sent_tf_vector
```

Menghitung nilai dari tf dan nilai dari idf

```python
tfidf_values1 = []
for token in word_tf_values1.keys():
    tfidf_sentences = [] # 1 dokumen adalah 1 kalimat
    for tf_sentence in word_tf_values1[token]:
        tf_idf_score = tf_sentence * word_idf_values1[token]
        tfidf_sentences.append(tf_idf_score)
    tfidf_values1.append(tfidf_sentences)
tfidf_values2 = []
for token2 in word_tf_values2.keys():
    tfidf_sentences2 = [] # 1 dokumen adalah 1 kalimat
    for tf_sentence2 in word_tf_values2[token2]:
        tf_idf_score2 = tf_sentence2 * word_idf_values2[token2]
        tfidf_sentences2.append(tf_idf_score2)
    tfidf_values2.append(tfidf_sentences2)
```
